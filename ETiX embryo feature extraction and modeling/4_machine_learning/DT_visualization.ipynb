```python
#plot decision tree and accuracy for each timepoint for defined subset of features df 

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import plot_tree

# Assuming your DataFrame is named df
# Adjust column names based on your actual DataFrame structure

# Get unique timepoints
timepoints = df['timepoint'].unique()

# Calculate the number of subplots needed
num_rows = len(timepoints) // 2 + len(timepoints) % 2
num_cols = 2

fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(15, 26), dpi=200)

for i, timepoint in enumerate(timepoints):
    # Subset the data for the current timepoint
    subset_df = df[df['timepoint'] == timepoint]
    
    # Separate features and target variable
    X = subset_df.drop(['embryo ID', 'timepoint', 'endpoint clusters'], axis=1)
    y = subset_df['endpoint clusters']

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)

    
    # Train a decision tree on the subset
    clf = DecisionTreeClassifier(max_depth=3, min_samples_leaf=5, random_state=0)
    clf.fit(X_train, y_train)
    
    # Make predictions on the subset
    y_pred = clf.predict(X_test)
    
    # Convert cluster numbers to strings
    class_names = list(map(str, sorted(df['endpoint clusters'].unique())))
    
    # Plot the decision tree
    ax = axes[i // 2, i % 2] if len(timepoints) > 1 else axes
    ax.set_title(f'Timepoint {timepoint}')
    plot_tree(clf, filled=True, feature_names=X.columns, class_names=class_names, ax=ax)
    
    # Print accuracy on the subset
    accuracy = accuracy_score(y_test, y_pred)
    ax.text(0.5, -0.1, f'Accuracy: {accuracy:.2f}, t= {timepoint}', size=8, ha="center", transform=ax.transAxes)

plt.tight_layout()
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import pandas as pd

# Assuming your DataFrame is named df
# Adjust column names based on your actual DataFrame structure

# Get unique timepoints
timepoints = df['timepoint'].unique()

# Initialize lists to store timepoints and corresponding accuracies
timepoint_list = []
accuracy_list = []

# Loop through each timepoint
for timepoint in timepoints:
    # Subset the data for the current timepoint
    subset_df = df[df['timepoint'] == timepoint]
    
    # Separate features and target variable
    X = subset_df.drop(["embryo ID", "timepoint", 'endpoint clusters'], axis=1)
    y = subset_df['endpoint clusters']
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)

    
    # Train a decision tree on the subset
    
    clf = DecisionTreeClassifier(max_depth=3, min_samples_leaf=5, random_state=0)
    clf.fit(X_train, y_train)
    
    # Make predictions on the subset
    y_pred = clf.predict(X_test)
  
    # Calculate accuracy for the current tree
    accuracy = accuracy_score(y_test, y_pred)
    
    # Store timepoint and accuracy in lists
    timepoint_list.append(timepoint)
    accuracy_list.append(accuracy)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(timepoint_list, accuracy_list, marker='o')
plt.xlabel('Timepoint')
plt.ylabel('Tree Accuracy')
plt.title('Decision Tree Accuracy per Timepoint')
plt.grid(True)
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np

# Assuming your DataFrame is named df
# Adjust column names based on your actual DataFrame structure

# Separate features (X) and target variable (y)
subset_df= df.copy()
subset_df= subset_df[subset_df['timepoint']==20]

X = subset_df.drop(["embryo ID", "timepoint", 'endpoint clusters'], axis=1)
y = subset_df['endpoint clusters']

# Define a range of max_leaf_nodes values to explore
max_leaf_nodes_range = range(2,15)

# Initialize lists to store max_leaf_nodes values and corresponding accuracies
max_leaf_nodes_list = []
accuracy_list = []

# Loop through different max_leaf_nodes values
for max_leaf_nodes in max_leaf_nodes_range:
    # Train a decision tree with the current max_leaf_nodes value
    clf = DecisionTreeClassifier(max_leaf_nodes=max_leaf_nodes)
    clf.fit(X, y)
    
    # Make predictions on the entire dataset
    y_pred = clf.predict(X)
    
    # Calculate accuracy for the current tree
    accuracy = accuracy_score(y, y_pred)
    
    # Store max_leaf_nodes and accuracy in lists
    max_leaf_nodes_list.append(max_leaf_nodes)
    accuracy_list.append(accuracy)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(max_leaf_nodes_list, accuracy_list, marker='o')
plt.xlabel('Max Leaf Nodes')
plt.ylabel('Tree Accuracy')
plt.title('Decision Tree Accuracy over Depth, t=20')
plt.grid(True)
plt.show()
```
